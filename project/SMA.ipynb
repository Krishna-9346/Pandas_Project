```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("social_media_usage.csv")

#Check null values
df.isnull().sum()

# Basic data overview
print("Dataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())

# Top platforms by average daily minutes spent
top_platforms = df.groupby("App")['Daily_Minutes_Spent'].mean().sort_values(ascending=False)
print("\nTop Platforms by Average Daily Minutes Spent:")
print(top_platforms)

# Correlation heatmap (excluding non-numeric columns)
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(8, 5))
cax = plt.matshow(corr_matrix, cmap='coolwarm')  #display the correlation matrix as a grid of colors.
plt.colorbar(cax)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45, ha='left')
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title("Correlation Heatmap of Social Media Usage")
plt.show()

# Scatter plot: Daily Minutes Spent vs. Posts Per Day
plt.figure(figsize=(8, 5))
colors = {'Facebook': 'blue', 'Instagram': 'purple', 'Twitter': 'cyan', 'Pinterest': 'red', 'Snapchat': 'orange'}
df['color'] = df['App'].map(colors).fillna('gray')  # Assign default color for NaN values
plt.scatter(df['Daily_Minutes_Spent'], df['Posts_Per_Day'], c=df['color'], alpha=0.7)
plt.xlabel("Daily Minutes Spent")
plt.ylabel("Posts Per Day")
plt.title("User Engagement: Time Spent vs. Posts Per Day")
plt.show()

# Bar plot: Average Likes per Day by Platform
plt.figure(figsize=(8, 5))
df.groupby("App")['Likes_Per_Day'].mean().sort_values().plot(kind='bar', color='purple')
plt.xlabel("App")
plt.ylabel("Average Likes Per Day")
plt.title("Average Likes Per Day by Platform")
plt.xticks(rotation=45)
plt.show()

# Histogram: Distribution of Daily Minutes Spent
plt.figure(figsize=(8, 5))
plt.hist(df['Daily_Minutes_Spent'].dropna(), bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel("Daily Minutes Spent")
plt.ylabel("Frequency")
plt.title("Distribution of Daily Minutes Spent on Social Media")
plt.show()

# Box Plot: Posts Per Day by Platform
plt.figure(figsize=(8, 5))
df.boxplot(column='Posts_Per_Day', by='App', patch_artist=True)
plt.xlabel("App")
plt.ylabel("Posts Per Day")
plt.title("Posts Per Day Distribution by Platform")
plt.xticks(rotation=45)
plt.show()

# Pie Chart: Platform Usage Share
plt.figure(figsize=(8, 5))
df['App'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='coolwarm', shadow=True)
plt.ylabel("")
plt.title("Platform Usage Share")
plt.show()
# Conclusion
print("\nInsights:")
print("- The platform with the highest daily minutes spent is:", top_platforms.idxmax())
print("- Strong correlation between time spent and likes received.")
```
